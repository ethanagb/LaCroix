### Example plots with matplotlib and seaborn 

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
lacroix.set_palette('Passionfruit_NoWhite')
sns.set_style("darkgrid")
for i in range(6):
    plt.plot(range(100), np.random.normal(i, 1, 100))
```

```python
berry = lacroix.colorList('Berry')
sns.palplot(sns.color_palette(berry))
```

```python
import seaborn as sns
sns.set(style="ticks")

dots = sns.load_dataset("dots")

# Define a palette to ensure that colors will be
# shared across the facets
palette = sns.color_palette(lacroix.colorList('CranRaspberry'))

# Plot the lines on two facets
sns.relplot(x="time", y="firing_rate",
            hue="coherence", size="choice", col="align",
            size_order=["T1", "T2"], palette=palette,
            height=5, aspect=.75, facet_kws=dict(sharex=False),
            kind="line", legend="full", data=dots)
```
